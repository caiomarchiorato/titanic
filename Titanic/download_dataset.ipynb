```python
from zipfile import ZipFile
import kaggle
import os
```

```python
from kaggle.api.kaggle_api_extended import KaggleApi
api = KaggleApi()
api.authenticate()

api.competition_download_files('titanic')

os.mkdir('./dados')
dados = "titanic.zip"
with ZipFile(dados, 'r') as zip:
    zip.printdir()
    zip.extractall('dados')
```
