```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv('dados/train.csv')
test = pd.read_csv('dados/test.csv')
```

```python
train.head()
test.head()
train.info()
```

```python
#criando o modelo que utilizará random forest que cria 100 modelinhos (arvores) e aplica para tomada de decisão
modelo = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
```

```python
#modelo só trabalha com variáveis numéricas, precisamos transformar VAR. que são nomes em números
#crianças e mulheres possuem uma proporção de salvamento maior (sex; age)
variaveis = ['Sex_binario', 'Age']

def transformar_sexo(valor):
    if valor == 'female':
        return 1
    else: 
        return 0

#map pega a funçao que eu passar e aplica linha a linha na coluna apresentado
train['Sex_binario'] = train['Sex'].map(transformar_sexo)
train.head()
```

```python
variaveis = ['Sex_binario', 'Age']
X = train[variaveis]
y = train['Survived']
#X.head()
#y.head()
X = X.fillna(-1)
#treinar o modelo
modelo.fit(X, y)
```

```python
#transformar as variáveis sex para binários também no test
test['Sex_binario'] = train['Sex'].map(transformar_sexo)
#test.head()
X_prev = test[variaveis]
X_prev = X_prev.fillna(-1)
#prevendo quais sobreviveram e quais não sobreviveram (array) com os dados de train e test
p = modelo.predict(X_prev)
p
```

```python
#transformando os arrays para uma série no pandas
pd.Series(p)
#alinhar a ordem de passageiros com as predições
sub = pd.Series(p, index=test['PassengerId'], name = 'Survived')
sub.shape
sub.to_csv("primeiro_modelo.csv", header= True)

sub.head()
```
